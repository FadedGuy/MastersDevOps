## Ejercicio 3

Define una función dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano

```python
import re
class Numero:
    def __init__(self, num: int):
        self.normal = num
        self.romano = self.toRomano(num)

    def imprime(self):
        print(str(self.normal) + " " + self.romano)

    def suma_romano(self, otroRomano):
        otroRomanoInt = self.fromRomano(otroRomano)
        self.normal += otroRomanoInt
        self.romano = self.toRomano(self.normal)

    def is_romano(self, strRomano: str):
        regexRomano = "^(M{0,3})(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$"

        return bool(re.match(regexRomano, strRomano))
    
    def toRomano(self, num):
        romans = [
            (1000, 'M'),
            (900, 'CM'),
            (500, 'D'),
            (400, 'CD'),
            (100, 'C'),
            (90, 'XC'),
            (50, 'L'),
            (40, 'XL'),
            (10, 'X'),
            (9, 'IX'),
            (5, 'V'),
            (4, 'IV'),
            (1, 'I')
        ]
        
        res = []
        for value, symbol in romans:
            if num == 0:
                break
            count = num // value
            res.append(symbol * count)
            num -= count * value

        return "".join(res)

    def fromRomano(self, str):
        numberDict = {
            "I": 1,
            "V": 5,
            "X": 10,
            "L": 50,
            "C": 100,
            "D": 500,
            "M": 1000
        }
        total = 0
        prev_valor = 0

        for char in reversed(str): 
            cur_valor = numberDict[char]
            if cur_valor < prev_valor:
                total -= cur_valor
            else:
                total += cur_valor
            prev_valor = cur_valor

        return total
```
